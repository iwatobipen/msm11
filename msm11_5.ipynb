```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
from sklearn import decomposition
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE
```

```python

iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
target_names = iris.target_names

pca = decomposition.PCA(n_components=2)
X_r = pca.fit(X).transform(X)

lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

tsne = TSNE(n_components=2, random_state=0)
X_r3 = tsne.fit_transform(iris.data)
```

```python
print('explained variance ratio (first two components): %s'
      % str(pca.explained_variance_ratio_))


fig, [ax1,ax2, ax3] = plt.subplots(1,3, figsize=(20, 6))

colors = ['navy', 'turquoise', 'darkorange']

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    ax1.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8,
                label=target_name)
ax1.legend(loc='best', shadow=False, scatterpoints=1)
ax1.set_title('PCA of IRIS dataset')

ax1.plot()


for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    ax2.scatter(X_r2[y == i, 0], X_r2[y == i, 1], alpha=.8, color=color,
                label=target_name)
ax2.legend(loc='best', shadow=False, scatterpoints=1)
ax2.set_title('LDA of IRIS dataset')

ax2.plot()

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    ax3.scatter(X_r3[y == i, 0], X_r3[y == i, 1], alpha=.8, color=color,
                label=target_name)
ax3.legend(loc='best', shadow=False, scatterpoints=1)
ax3.set_title('tSNE of IRIS dataset')

ax3.plot()
```
